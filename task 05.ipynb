```python
df.drop(['Owner_of_vehicle', 'Casualty_class','Vehicle_driver_relation', 'Sex_of_casualty', 'Age_band_of_casualty', 'Work_of_casuality', 'Fitness_of_casuality', 'Pedestrian_movement', 'Casualty_severity'], axis=1, inplace=True)
```

```python
for i in col:
    if df[i].isnull().sum() > 0:
        mode_value = df[i].mode()[0]  # Calculate mode for the column
        df[i].fillna(mode_value, inplace=True)  # Fill missing values with mode
    else:
        continue
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = df['Time'].dt.hour
```

```python
df.drop('Time', axis=1, inplace=True)
```

```python
df.columns
```

```python
sns.lineplot(x=df['Hour'], y=df['Hour'].value_counts())
plt.xlabel('Hour')
plt.ylabel('Numebr of Accidents')
plt.title('Peak Accident Hours',color= 'red')
plt.show()
```

#### Observation: There are more accidents between 6 am and 9 am and at midnight.

```python
sns.countplot(x='Sex_of_driver', data=df)
plt.xlabel('Sex of Driver')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Sex of Driver', color='red')
plt.show()
```

#### Observation : Men are more involved in accidents as compared to women

```python
sns.countplot(x='Age_band_of_driver', data=df)
plt.xlabel('Age band of driver')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Age of Driver', color='red')
plt.show()
```

#### Observation : Age 18-50 people are more engaged in accidents.

```python
plt.figure(figsize=(8,8))  # Set the size of the figure
sns.set_palette('pastel')  # Set color palette
plt.pie(df['Type_of_vehicle'].value_counts(), labels=df['Type_of_vehicle'].value_counts().index, autopct='%1.0f%%', startangle=90)
plt.axis('equal')  

# Add a title
plt.title('Pie Chart Example')

# Show the plot
plt.show()
```

#### Observation : Automobiles and Lorries are more likely to confront with accident

```python
sns.countplot(y='Area_accident_occured', data=df)
plt.xlabel('Area of Accident')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Area', color='red')
plt.show()
```

#### Observation : Residential and Office Areas are hotspots for accidents

```python
sns.countplot(x='Weather_conditions', hue='Accident_severity', data=df)
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.title('Accident Severity by Weather Conditions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Accident Severity')
plt.show()
```

#### Observations : Slight Injuries are very likely to be caused in normal and rainy weather.
   ####             Serious and fatal injuries are very rare. 

```python
sns.countplot(y='Type_of_collision', data=df)
plt.xlabel('Number of Accidents')
plt.ylabel('Type of collision')
plt.title('Distribution of Accidents by Type of Collision', color='red')
plt.show()
```

#### Observation : Vehicle to Vehicle collision is very common.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Cause_of_accident', data=df)
plt.xlabel('Numebr of Accident')
plt.ylabel('Cause of Accidents')
plt.title('Number of Accidents by Cause of Accident')
plt.tight_layout()
plt.show()
```

### Observations:
 ####   1. Safe distance is crutial to avoid accident.
 ####   2. Changing lane is most frequent act for accident.
 ####   3. Many accidents are caused when vehicle is being reversed.
